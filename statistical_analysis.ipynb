```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
from pylab import rcParams
from scipy.stats import f_oneway
from scipy.stats import ttest_ind
%matplotlib inline
warnings.filterwarnings("ignore")
rcParams['figure.figsize'] = 20,10
rcParams['font.size'] = 30
sns.set()
np.random.seed(8)
```

```python
with_syncends = np.array([60,67,65,63,59,56,65,63,67,78,62,65,58,67,59,58,68,65,62,69,58,68,58,61,63,67,62,59,63,62,52,59,63,64,62,53,56,63,62,68,59,63,70,65,69,55,60,69,62,67])
without_syncends = np.array([110,134,230,168,150,149,178,290,189,148,137,167,250,176,133,148,189,230,130,144,188,183,235,176,123,117,123,195,126,144,119,123,175,125,118,129,143,235,187,138,124,168,215,156,103,144,194,220,189,143])
```

```python
def plot_distribution(inp):
    plt.figure()
    ax = sns.distplot(inp)
    plt.axvline(np.mean(inp), color="k", linestyle="dashed", linewidth=5)
    _, max_ = plt.ylim()
    plt.text(
        inp.mean() + inp.mean() / 10,
        max_ - max_ / 10,
        "Mean: {:.2f}".format(inp.mean()),
    )
    return plt.figure
```

```python
plot_distribution(with_syncends)
plot_distribution(without_syncends)
```

```python
plt.figure()
ax1 = sns.distplot(with_syncends)
ax2 = sns.distplot(without_syncends)
plt.axvline(np.mean(with_syncends), color='b', linestyle='dashed', linewidth=5)
plt.axvline(np.mean(without_syncends), color='orange', linestyle='dashed', linewidth=5)
```

```python
def compare_2_groups(arr_1, arr_2, alpha, sample_size):
    stat, p = ttest_ind(arr_1, arr_2)
    print('Statistics=%.3f, p=%.3f' % (stat, p))
    if p > alpha:
        print('Same distributions (fail to reject H0)')
    else:
        print('Different distributions (reject H0)')
```

```python
sample_size = 25
with_syncends_sampled = np.random.choice(with_syncends, sample_size)
without_syncends_sampled = np.random.choice(without_syncends, sample_size)
compare_2_groups(with_syncends_sampled, without_syncends_sampled, 0.05, sample_size)
```
